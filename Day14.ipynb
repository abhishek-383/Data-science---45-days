# **Coloum add & rename**

```python
import pandas as pd
d = {
    "name":["vishal","virat","vineet"], # 3 rows
    "salary":[100,200,300] # 3 rows
}
df = pd.DataFrame(data=d)
# add single column
df["marks"] = [10,20,30]
df["marks"] = df["salary"] / 2
df["increment"] = df["salary"] + (df["salary"] / 10)
# column name change
df.rename(columns={"increment":"salary_increment"},inplace=True)
df.drop(["salary","name"],axis=1,inplace=True)
df
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/PraveenBunker/45-days-of-python/main/student-data1.json"
df  = pd.read_json(url)
print(df)
print(df.iloc[0:6,[0,1]])
print(df.loc[1:5,"name"])
print(df.iloc[[3,8],[0]])
print(df.iloc[3:9:5,0])
# filter

df[df["english"] == 95]

df[df["maths"] > 60]

#multiple conditions
print(df[(df["english"] > 60) & (df["maths"] > 60)]["name"])
print(df[(df["english"] > 60) | (df["maths"] > 60)]["name"])
```
