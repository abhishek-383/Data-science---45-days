# **Student data manager**
# Using oops
# Features : add, update, delete, search

```python
class Student:
    def __init__(self, student_id, name, age, course):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.course = course

    def display(self):
        print(f"ID: {self.student_id}, Name: {self.name}, Age: {self.age}, Course: {self.course}")


class StudentManager:
    def __init__(self):
        self.students = {}

    # Add Student
    def add_student(self, student):
        if student.student_id in self.students:
            print("Student ID already exists!")
        else:
            self.students[student.student_id] = student
            print("Student added successfully.")

    # Update Student
    def update_student(self, student_id):
        if student_id in self.students:
            student = self.students[student_id]

            print("\nEnter new details:")
            student.name = input("Name: ")
            student.age = int(input("Age: "))
            student.course = input("Course: ")

            print("Student record updated successfully.")
        else:
            print("Student not found!")

    # Delete Student
    def delete_student(self, student_id):
        if student_id in self.students:
            del self.students[student_id]
            print("Student deleted successfully.")
        else:
            print("Student not found!")

    # Search Student
    def search_student(self, student_id):
        if student_id in self.students:
            print("\nStudent Found:")
            self.students[student_id].display()
        else:
            print("Student not found!")

    # Display All Students
    def display_all(self):
        if not self.students:
            print("No student records available.")
        else:
            print("\nStudent Records:")
            for student in self.students.values():
                student.display()


manager = StudentManager()

while True:
    print("|< Student Data Manager >|")
    print("     1. Add Student")
    print("     2. Update Student")
    print("     3. Delete Student")
    print("     4. Search Student")
    print("     5. Display All Students")
    print("     6. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        sid = input("Enter Student ID: ")
        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        course = input("Enter Course: ")

        student = Student(sid, name, age, course)
        manager.add_student(student)

    elif choice == "2":
        sid = input("Enter Student ID to update: ")
        manager.update_student(sid)

    elif choice == "3":
        sid = input("Enter Student ID to delete: ")
        manager.delete_student(sid)

    elif choice == "4":
        sid = input("Enter Student ID to search: ")
        manager.search_student(sid)

    elif choice == "5":
        manager.display_all()

    elif choice == "6":
        print("Byeeee")
        break

    else:
        print("Invalid choice")
```
