# *Importing and using modules*

```python
#import modules
import math
print(math.factorial(4))
print(math.pow(4,2))
print(math.pi)
print(math.floor(10.7234))
print(math.ceil(10.0123))
```

```python
#ex 1
import math as AJ
print(AJ.factorial(4))
print(AJ.pow(4,2))
print(AJ.pi)
print(AJ.floor(10.7234))
print(AJ.ceil(10.0123))
```

```python
#ex 2
from math import factorial,pow
print(factorial(4))
print(pow(4,2))
```

```python
#ex 3
from math import *
print(factorial(4))
print(pow(4,2))
print(pi)
```
