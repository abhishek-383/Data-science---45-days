# **Histogram**

```python
import matplotlib.pyplot as plt
marks =[40,55,60,70,75,90,33,50]
plt.hist(marks,bins=2,color="green")
plt.show()
```

# **Pie chart**

```python
# pie chart
import matplotlib.pyplot as plt
fruits = ["apple","banana","orange","watermelon"]
count =[40,30,15,70]
plt.figure(figsize=(4,3))
plt.pie(count,labels=fruits)
plt.show()
```

```python
import matplotlib.pyplot as plt
fruits = ["apple","banana","orange","watermelon"]
count =[40,30,15,70]
plt.pie(count,labels=fruits,startangle=270,autopct="%1.1f%%")   #  f is for float  | we can use i for integer

plt.show()
```

```python
# use of subplot

import matplotlib.pyplot as plt
# first chart
x= [1,2,3,4,5]
y = [10,20,30,40,55]
plt.subplot(1,2,1)
plt.plot(x,y)

plt.xlabel("x axis")
plt.ylabel("y axis")

# second chart

x1 = ["apple","banana","orange","watermelon"]
y1 =[40,30,15,70]
plt.subplot(1,2,2)
plt.pie(y1,labels=x1,startangle=80)
plt.xlabel("x1 axis")
plt.ylabel("y1 axis")

# plt.legend()
plt.title("pie chart")

plt.tight_layout()
plt.show()
```

# **Subplots**

```python
import matplotlib.pyplot as plt
# first chart
x= [1,2,3,4,5]
y = [10,20,30,40,55]
plt.subplot(2,2,1)
plt.plot(x,y)

plt.xlabel("x axis")
plt.ylabel("y axis")

# second chart

x1 = ["apple","banana","orange","watermelon"]
y1 =[40,30,15,70]
plt.subplot(2,2,2)
plt.pie(y1,labels=x1,startangle=80)
plt.xlabel("x1 axis")
plt.ylabel("y1 axis")

# plt.legend()
plt.title("pie chart")

#third chart
marks =[40,55,60,70,75,90,33,50]
plt.subplot(2,2,3)
plt.xlabel("marks")
plt.ylabel("count")
plt.title("histogram")
plt.hist(marks,bins=2,color="green")


# fourth chart

fruits = ["apple","banana","orange","watermelon"]
count =[40,30,15,70]
plt.subplot(2,2,4)
plt.figure(figsize=(4,3))
plt.pie(count,labels=fruits)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))

# First chart
x = [1,2,3,4,5]
y = [10,20,30,40,55]
plt.subplot(2,2,1)
plt.plot(x, y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Chart")

# Second chart
x1 = ["apple","banana","orange","watermelon"]
y1 = [40,30,15,70]
plt.subplot(2,2,2)
plt.pie(y1, labels=x1, startangle=80)
plt.title("Pie Chart")

# Third chart
marks = [40,55,60,70,75,90,33,50]
plt.subplot(2,2,3)
plt.hist(marks, bins=2, color="green")
plt.xlabel("marks")
plt.ylabel("count")
plt.title("Histogram")

# Fourth chart
fruits = ["apple","banana","orange","watermelon"]
count = [40,30,15,70]
plt.subplot(2,2,4)
plt.pie(count, labels=fruits)
plt.title("Pie Chart 2")

plt.tight_layout()
plt.show()
```
