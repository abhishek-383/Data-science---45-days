**Sorting**
1. df.sort_values("column_name") -> Sort by Single Column
2. df.sort_values("Age", ascending=True) -> Sort in Ascending Order
3. df.sort_values("Age", ascending=False) -> Sort in Descending Order
4. df.sort_values(["col1", "col2"]) -> Sort Multiple Columns
5. df.sort_values( ["Age", "Salary"], ascending=[True, False] ) -> Multiple Columns with Descending Order

```python
import pandas as pd
url = "https://raw.githubusercontent.com/PraveenBunker/45-days-of-python/main/student-data1.json"
df  = pd.read_json(url)
df.sort_values("english")    #  by default ascending
df.sort_values("name",ascending=[False])  # descending

df.sort_values(by=["english"],ascending=False)
df.sort_values(by=["english","maths"],ascending=[False,True])

df["name"] = df["name"].str.lower()
print(df.sort_values(by=["name"],ascending=[True]))
```

### **Working with date** **value**
1. df['doj'] -> Add New Date Column
2. df['doj'].dtype -> Check Data Type of Date Column
3. df['doj'] = pd.to_datetime(df['doj']) -> Convert Object to Datetime
4. df['doj'].dtype -> Check Data Type of Date Column
5. pd.Timestamp.now() -> Get Current Date & Time
6. df['doj'].dt.year -> Extract Year
7. df['doj'].dt.month -> Extract Month
8. df['doj'].dt.day -> Extract Day
9. df[(df['doj'] >= '2024-01-01') & (df['doj'] <= '2024-12-31')] -> Filter Between Two Dates

```python

import pandas as pd
url = "https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/file2.json"
df = pd.read_json(url)
df['doj'] = ['2025-01-01',"2025-02-02","2025-03-03","2025-04-04","2025-05-05"]
df
df['doj'].dtype   # output ('O') it means string type
df['doj'] = pd.to_datetime(df['doj'])
df['doj'].dtype # output ('<M8[ns]') it means date formate

# Date operation
df['doj'].dt.year
df['doj'].dt.month
df['doj'].dt.day
df['doj'].dt.day_name()
df['doj'].dt.is_leap_year

#add 20 days
# df['doj'] + pd.Timedelta(days=20)
df['doj'] + pd.to_timedelta("20 days")

```
