## Ex of logical operators

# Bitwise operator

```python
# >> , << , & ,
 #Bitwise right shift (>>)
a = 10  # Binary: 1010
b = a >> 2 # Shifts 2 bits to the right
print(f"10 >> 2 = {b}") # Expected: 2 (Binary: 0010)

# Bitwise left shift (<<)
c = 5   # Binary: 0101
d = c << 1 # Shifts 1 bit to the left
print(f"5 << 1 = {d}")   # Expected: 10 (Binary: 1010)

# Bitwise AND (&)
e = 12  # Binary: 1100
f = 25  # Binary: 11001
g = e & f
print(f"12 & 25 = {g}") # Expected: 8 (Binary: 01000)

# Bitwise OR (|)
h = 12  # Binary: 1100
i = 25  # Binary: 11001
j = h | i
print(f"12 | 25 = {j}") # Expected: 29 (Binary: 11101)

# Bitwise XOR (^)
k = 12  # Binary: 1100
l = 25  # Binary: 11001
m = k ^ l
print(f"12 ^ 25 = {m}") # Expected: 21 (Binary: 10101)

# Bitwise NOT (~)
n = 10  # Binary: 0...01010 (assuming 2's complement)
o = ~n
print(f"~10 = {o}")    # Expected: -11
```
