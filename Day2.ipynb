# operator & expression

```python
#arthmetic opr = + - * / // % **

a = 30
b = 20
print(a+b) #addition

print(a-b) #sub

print(a*b) #multiply

print(a/b) #division (output in float)

print(a//b) #float division (outout in integer)

print(a%b) #modulas (remainder)

a=5
b=3
print(a**b) #double star (for power)
```

```python
#Relational opr -> output in true and false
# (== ,< , > , >= , <= , != )

# ==
a=10
b=10
print(a==b)

# < , >
a=11
b=12
print(a<b)
print(a>b)

# <=
a=10
b=10
print(a>=b)

# >=
a=20
b=10
print(a<=b)

#!=
a=10
b=20
print(a!=b)
```

```python
# assignment operator
# = , += , -=

# =
a=10
print(a)

# +=
a+=10
print(a)

# -=
a-=10
print(a)
```

```python
# Logical operator
# And , or , not

a=""
b="HELLO"
print(a and b) #output is b
print(b and a) #output is a
print(a or b) #output is a
print(b or a) #output is b
print(not a) #output is false
print(not 0) #output is true
```
