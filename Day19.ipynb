# **Subplots**

```python
# Subplots
import matplotlib.pyplot as plt
# graph one data
year = [2010,2015,2020,2025]
dairy = [100,520,630,400]

# graph two data
year = [1990,2000,2005,2010]
farming = [300,200,250,100]

fig , aux = plt.subplots(1,2)
# for first graph
aux[0].plot(year,dairy)
aux[0].set_xlabel('year')
aux[0].set_ylabel('dairy')
aux[0].set_title('dairy production graph')

# for second graph
aux[1].plot(year,farming)
aux[1].set_xlabel('year')
aux[1].set_ylabel('farming')
aux[1].set_title('farming production graph')
plt.tight_layout()
plt.show()
```

# **Subplots for 2d**

```python
import matplotlib.pyplot as plt

fig, aux = plt.subplots(2, 2)

# Graph 1
year1 = [2010, 2015, 2020, 2025]
dairy = [100, 520, 630, 400]

aux[0][0].plot(year1, dairy)
aux[0][0].set_xlabel("Year")
aux[0][0].set_ylabel("Dairy")
aux[0][0].set_title("Dairy Production")

# Graph 2
year2 = [1990, 2000, 2005, 2010]
farming = [300, 200, 250, 100]

aux[0][1].plot(year2, farming)
aux[0][1].set_xlabel("Year")
aux[0][1].set_ylabel("Farming")
aux[0][1].set_title("Farming Production")

# Graph 3
year3 = [2012, 2016, 2020, 2024]
milk = [150, 250, 350, 450]

aux[1][0].plot(year3, milk)
aux[1][0].set_xlabel("Year")
aux[1][0].set_ylabel("Milk")
aux[1][0].set_title("Milk Production")

# Graph 4
year4 = [2011, 2014, 2018, 2022]
crop = [500, 450, 600, 700]

aux[1][1].plot(year4, crop)
aux[1][1].set_xlabel("Year")
aux[1][1].set_ylabel("Crop")
aux[1][1].set_title("Crop Production")

plt.tight_layout()
# plt.show()
plt.gcf().canvas.get_supported_filetypes()
plt.savefig('subplots.jpg')
plt.show()
```
