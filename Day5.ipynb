```python
d= { "name":"hello"}
d.update({"name":"Arya mains"})
print(d)
```

```python
d= { "name":"hello","age":"13"}
d.popitem()
print(d)
```

```python
d = { "name": "hello", "age": "13", "city": "New York" }

# 1. get() method
print("Using get():", d.get("name"))
print("Using get() for non-existent key with default value:", d.get("country", "USA"))

# 2. keys() method
print("Keys:", d.keys())

# 3. values() method
print("Values:", d.values())

# 4. copy() method
d_copy = d.copy()
print("Copied dictionary:", d_copy)
d_copy["city"] = "Los Angeles"
print("Original dictionary after modifying copy:", d)
print("Modified copied dictionary:", d_copy)

# 5. del keyword
del d["name"]
print("Dictionary after deleting 'name':", d)

# 6. clear() method
d.clear()
print("Dictionary after clear():", d)
```

# Nested list

```python
l = [10,20,30,["hello","hello1","hello2",[True,False]]]
print(l[0])
print(l[1])
print(l[2])
print(l[3][0])
print(l[3][1])
print(l[3][2])
print(l[3][3][0])
print(l[3][3][1])
#2nd method
for i in range(len(l)-1):
    print(l[i])
for j in range(len(l[3])-1):
    print(l[3][j])
for k in range(len(l[3][3])):
    print(l[3][3][k])
```

# Slicing list

```python
l = [10,20,30,["hello","hello1","hello2",[True,False]]]
l1 = l[1:]
print(l1)
l1 = l[:3]
print(l1)
l1 = l[1:3]
print(l1)
print(l[3][1:3])
```

# 1. Map (transform values), filter(select values),reduce(combine values)
#2. importing & using modules

```python
#minimum 2 arguments needed to use these following methods

def square(x):
  return x*x

#map
l=[10,20,30]
l1 = list(map(square,l))
print(l1)

l2 = list(map(lambda x:x*x,l))
print(l2)

#alternative | without in built function
l1 = []
for i in range(len(l)):
  l1.append(l[i]*l[i])
print(l1)

#filter
def helo(x):
  return x.endswith("a")
l = ["apple","banana","cat","dog"]
l1 = list(filter(helo,l))
print(l1)

l2 = list(filter(lambda x:x.endswith("t"),l))
print(l2)

#alternative

#dsa wise
l1=[]
for i in l:
  if 'g' == i[-1]:
    l1.append(i)
print(l1)

#reduce

numbers = [1, 2, 3, 4, 5]

# Sum all numbers
result = reduce(lambda x, y: x + y, numbers)

print(result)
```

# Exception handling


```python
# try exception
try:
  num1 = 10
  num2 = 5
  print(num1/num2)
except:
  print("hello except")

try:
  num1 = 10
  num2 = 0
  print(num1/num2)
except:
  print("hello except")

#try except & final
try:
  num1 = 10
  num2 = 3
  print(num1/num2)
except:
  print("hello except")

finally:     #always executed error occure or not occure
  print("work done")
```
