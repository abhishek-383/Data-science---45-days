## Lambda function(lambda arguments : expression)

```python
sum = lambda x,y : x+y
print(sum(10,20))

square = lambda x : x*x
print(square(10))

```

```python
#Question

show = lambda *a : print(a)
show(10,20,30,40)
show("AJ","Saini")
```

# List comprihensions

```python
l = [10,20,[30,40,50,60]]
print([l[-1][i] for i in range(len(l[2]))])
squares = [(x * x) for x in range(5)]
print(squares)

words = ["python", "java", "c++"]

upper_words = [i.upper() for i in words]
print(upper_words)
```
