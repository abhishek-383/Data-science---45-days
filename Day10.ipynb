# **Creating array from list**

```python
#1d list
l = [1,2,3]
print(l)

#1d array
import numpy as np
arr = np.array(l)
print(arr)

#2d list

l = [[1,2,3],[4,5,6]]
print(l)
l[1][0]= 100
print(l)

#2d array
n = np.array([[1,2,3],[4,5,6]])
print(n)
print(n[0]) #1st row -> [1,2,3]

```

# **list vs array**
    1. Multiply
    2. Comparasion
    3. Zeros
    4. Ones
    5. Full
    6. Random
    7. Arrange

```python
# list
l = [1,2,3]
lm = l*2
print(lm)

#array
import numpy as np
arr = np.array(l)
arrm = arr*2
print(arrm)
```

```python
#comparasion
#  array takes less time comparasion by list
#list
import time
start = time.time()
l = [i*500 for i in range(1000000)]
print("list output:" , time.time() - start)


#array
import numpy as np
start = time.time()
arr = np.arange(1000000)*500
print("array output:" , time.time() - start)
```

```python
# Zeros array
import numpy as np
arr = np.zeros(5)
print(arr)
# 2d
arr = np.zeros((2,5))
print(arr)
```

```python
# ones array 1d
import numpy as np
arr = np.ones(5)
print(arr)
#ones array 2d
arr = np.ones((2,5))
print(arr)
```

```python
# Question
# use of zeros increase by 10
# use of ones increase 5
import numpy as np
arr = np.zeros((2,5))
print(arr+10)
arr = np.ones((2,5))
print(arr*5)
```

```python
# full for 1d
import numpy as np
arr = np.full(5,8)
print(arr)
# full for 2d
arr = np.full((2,6),7)
print(arr)
```

```python
# Question
# by the help of zeros -> 2d -> 6
# by the help of full -> 2d -> 1
import numpy as np
arr = np.zeros((2,6))+6
print(arr)
arr = np.full((2,6),1)
print(arr)
```

```python
# Random
import random
import numpy as np
arr = np.random.random((2,6))
print(arr)
```

```python
# arange for 1d
import numpy as np
arr = np.arange(5)
print(arr)
# arange for 2d
arr = np.arange(20).reshape(4,5)
print(arr)
```
