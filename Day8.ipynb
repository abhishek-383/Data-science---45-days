# Inheritance

```python
class Address:
  def __init__(self,city,state):
    self.city=city
    self.state=state

  def location(self):
    return f"My location is {self.city} in {self.state}"

class User(Address):
  def __init__(self,name,age,city,state):
    self.name=name
    self.age=age
    super().__init__(city,state) #super keyword to remove below two lines
    # self.city=city
    # self.state=state #
  def userDetails(self):
    print(f"My name is {self.name}")
    print(f"My age is {self.age}")
    print(super().location())


u= User("aj",20,"jaipur","rajasthan")
print(u.city)
u.userDetails()
```

# **Encapsulatin**

```python
class Address:
  def __init__(self,city,state):
    self.__city=city  #private attribute (start with double underscore) to access we create a new f^n
    self.state=state

  def locatin(self):
    print(f"My location is {self.__city} in {self.state}")

  def getcity(self):
    print(self.__city)

a = Address("Chhapara","bihar")
a.locatin()
a.getcity()
print(a.state)
```

**Polymorphism**

```python
class Address:
  def __init__(self,city,state):
    self.city=city
    self.state=state

  def location(self):
    print(f"My current location is {self.city} in {self.state}")

class hometown:
  def __init__(self,city,state):
    self.city=city
    self.state=state

  def location(self):
    print(f"My hometown location is {self.city} in {self.state}")
a = Address("Vadodara","gujrat")
a.location()
b = hometown("jaipur","rajasthan")
b.location()
```

# Overloading & **Overriding**

```python
def Address(city,state):
  print(f"My address is {city} in {state} ")

def Address(city,state,country):
  print(f"My address is {city} at {state} in {country} ")

# error -> Address("jaipur","rajasthan")    [Any type of overloading is not possible in python]
Address("jaipur","rajasthan","india")
```

```python
#overriding
class Address:
  def __init__(self,city,state):
    self.city=city
    self.state=state

  def location(self):
    print(f"Address  ")

class user(Address):
  def __init__(self,name,age,city,state):
    super().__init__(city,state)
    self.name=name
    self.age=age

  def location(self):               # if we comment these both lines output is Address
    print(f"user location ")

u = user("arya main", 20,"kukas","rajasthan")
u.location()
```
