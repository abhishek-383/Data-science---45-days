Array reshaping
 1. reshape()
 2. flatten() -> create copy
 3. ravel() - > return original
 4. transpose -> using T

```python
# Reshape array
import numpy as np
arr = np.arange(12)
up_arr = arr.reshape(3,4) # 3 -> rows | 4 -> coloums
print(up_arr)
```

```python
# example
import numpy as np
arr = np.arange(16).reshape(2,4,2) # 2 -> array | 4 -> rows in both | 2 -> coloums in both
print(arr)

```

```python
# flatten -> create a copy than works
import numpy as np
arr = np.arange(12).reshape(3,4)
print(arr)
up_arr = arr.flatten()
print(up_arr)
print(arr)

# exanple for 3d
arr1 = np.arange(16).reshape(2,4,2)
print(arr1)
up_arr1 = arr1.flatten()
up_arr1[0] = 100
print(up_arr1)
print(arr1)
```

```python
# Ravel for 2d -> works on original array
import numpy as np
arr = np.arange(14).reshape(2,7)
print(arr)
up_arr = arr.ravel()
print(up_arr)
print(arr)

# ravel for 3d
arr1 = np.arange(32).reshape(2,4,4)
print(arr1)
up_arr1 = arr1.ravel()
up_arr1[0] = 100
print(up_arr1)
print(arr1)
```

```python
# Transpose -> convert row into coloum

import numpy as np
arr = np.arange(16).reshape(4,4)
print(arr)
up_arr = arr.T
print(up_arr)
arr1 = np.arange(16)
print(arr1)
up_arr = arr1.reshape(4,4).T
print(up_arr)

# for 3d
arr2 = np.arange(32).reshape(2,4,4)
print(arr2)
up_arr2 = arr2.T
print(up_arr2)
```

# Slicing for 1d 2d 3d

```python
# for 1d
import numpy as np
arr = np.arange(10)
print(arr[3:6])
print(arr[:6])
print(arr[3:])
print(arr[3:6:2])
print(arr)
```

```python
# for 2d
import numpy as np
arr = np.arange(16).reshape(2,8)
print(arr)
up_arr = arr[:,-1]
print(up_arr)
```

```python
# for 3d
import numpy as np
arr = np.arange(24).reshape(3,4,2)
print(arr)
up_arr = arr[::2,::3,::]
print(arr[0::2,0::3,0::1][:,[0,1],[0,1]])
# print(up_arr)
```

Looping using Numpy
  1. while loop
  2. for loop

```python
# while loop
# for 1d
import numpy as np
arr = np.arange(12)
i = 0
while i< len(arr):
  print(arr[i], end=" ")
  i += 1


# for 2d
import numpy as np
arr = np.arange(16).reshape(4,4)
i = 0
while i< len(arr):
  j = 0
  while j< len(arr[i]):
    print(arr[i][j], end=" ")
    j += 1

  print("\n")
  i += 1


```

```python
# for loop
# for 1d
import numpy as np
arr = np.arange(12)
for i in range(12):
  print(arr[i], end="  ")

arr1 = np.arange(16).reshape(2,8)
for i in range(len(arr1)):
  for j in range(len(arr1[i])):
    print(arr1[i][j], end=" ")
  print("\n")
```

```python
import numpy as np
arr = np.zeros(16).reshape(4,4)
for i in range(len(arr)):
  for j in range(len(arr[i])):
    if i == j:
      arr[i][j] = 1
    print(arr[i][j], end=" ")
  print("\n")
```
