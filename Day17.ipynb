# **Multiple** **Line Chart**

```python
import matplotlib.pyplot as plt
x= [1, 2, 3, 4]
y1 = [100, 120, 115, 140]
y2: [90, 110, 130, 125]
plt.plot(x, y1, label='jeans')
plt.plot(x, y2, label='shirt')
plt.plot(color="m",marker='*',linestyle=':',linewidth= 2,markersize=10)
plt.legend()  # info of label
plt.show()
```

```python
# multi lines chart
import matplotlib.pyplot as plt
x=[1,2,3,4]
y1 = [100,200,260,290]
y2 = [155,185,195,300]
plt.plot(x,y1,y2,marker='*',linestyle=':',linewidth= 2,markersize=10)
plt.xlabel("Months")
plt.ylabel("Store")
plt.title("sales Report")
plt.legend()  # info of label
plt.show()
```

# Multiple **chart**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y1 = [10,20,20,40]
y2= [20,30,25,30]

# calculation -> width
w = 0.20

plt.figure(figsize=(4,3))
plt.bar(x-w/2,y1,label="jeans",width=w)
plt.bar(x+w/2,y2,label="shirt",width=w)
plt.xlabel("Groups")
plt.ylabel("Number of students")
plt.title("Number of students in each group")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.array([1,2,3,4])
y1 = [20,30,25,40]
y2 = [10,15,25,45]
y3 = [5,35,20,10]

w= 0.70
# plt.figure(figsize=(4,3))
plt.bar(x-w/3,y1,label="sssss",width=w/3)
plt.bar(x,y2,label="jeans",width=w/3)
plt.bar(x+w/3,y3,label="shirt",width=w/3)
plt.xlabel("Groups")
plt.ylabel("Number of students")
plt.title("Number of students in each group")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.array([1,2,3,4])
y1 = [20,30,25,40]
y2 = [10,15,25,45]
y3 = [5,35,20,10]
y4 = [2,44,33,22]

w= 0.40
# plt.figure(figsize=(4,3))
plt.bar(x-0.50,y1,label="sssss",width=0.25)
plt.bar(x-0.25,y2,label="jeans",width=0.25)
plt.bar(x+0.25,y3,label="shirt",width=0.25)
plt.bar(x+0.50,y4,label="e",width=0.25)
plt.xlabel("Groups")
plt.ylabel("Number of students")
plt.title("Number of students in each group")
plt.legend()
plt.show()
```
